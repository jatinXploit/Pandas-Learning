```python
import numpy as np
import pandas as pd 
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('Category').sum('sales')
```

```python
cat
```

```python
cat = df.groupby('Category')
cat
for i,v in cat:
       print(i)
       print(v)
```

```python
# Group by Category and calculate the sum of Sale
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by Store and calculate the sum of Sales
store = df.groupby('Store')['Sales'].sum()
store
```

```python
# Group by multiple columns
#(or)
# Group by Category and Store
df.groupby(['Category','Store'])['Sales'].sum()
```
