merging

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    'employee_id' : [1, 2, 3, 4, 5],
    'name' : ['jaitn', 'mayank', 'bhartee', 'shashank','supriya'],
    'Department' : ['IT', 'HR','HR', 'IT', 'HR']
})
salary = pd.DataFrame({
    'employee_id' : [1, 2, 3, 6, 7],
    'salary' : [ 200000, 150000, 200000, 150000,150000],
    'bonus' : [ 20000, 30000, 40000, 20000, 30000]
})
```

```python
employees
```

```python
salary
```

```python
pd.merge(employees, salary)
```

```python
pd.merge(employees, salary, on = 'employee_id', how = 'inner')
```

```python
pd.merge(employees, salary, on = 'employee_id', how = 'outer')
```

```python
pd.merge(employees, salary, on = 'employee_id', how = 'left')
```

```python
pd.merge(employees, salary, on = 'employee_id', how = 'right')
```

CONCATINATION

```python
df1 = pd.DataFrame({
    'A' : [ 'A0', 'A1', 'A2'],
    'B' : [ 'B0', 'B1', 'B2'],
    'C' : [ 'C1', 'C2', 'C3']
})

df2 = pd.DataFrame({
    'A' : [ 'A3', 'A4', 'A5'],
    'B' : [ 'B3', 'B4', 'B5'],
    'C' : [ 'C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 0)
```

```python
pd.concat([df1, df2], axis = 1)
```

JOINING

```python
df1 = pd.DataFrame({
    'name': ['jatin', 'mayank', 'bhartee', 'supriya']
},index = [1, 2, 3, 4])
df2 = pd.DataFrame({
    'score': [ 99, 98, 97, 96]
}, index = [2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2)
```
